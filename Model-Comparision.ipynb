<h1><b> ROC-AUC

```python
categories = ['Logistic\nRegression', 'Decision\nTree', 'Random\nForest', 'SVM', 'XGBoost']
values = [
    Log_reg_ROC_AUC_Score,
    Dec_tree_ROC_AUC_Score,
    random_forest_ROC_AUC_Score,
    svm_ROC_AUC_Score,
    XGBoost_ROC_AUC_Score
]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors, edgecolor='black')
plt.xlabel('Models')
plt.ylabel('ROC AUC Score')
plt.title('Model Comparison')

for i, v in enumerate(values):
    plt.text(i, v + 0.01, f'{v:.2f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
